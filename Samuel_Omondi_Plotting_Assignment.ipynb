**TO DO**: The dataset has a number of different statistics that describe fuel efficiency. Here, create a _histogram_ that depicts the distribution of carbon Dioxide values taken. Try playing around with different bin width sizes to see what best depicts the data. Define if the distribution is skewed or symmetric.

Make sure you create 2 more histograms based on columns of your choice

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/fuel_econ.csv')

# 1. Histogram for `co2` - carbon Dioxide values
plt.figure(figsize=(10, 6))
plt.hist(data['co2'], bins=50, edgecolor='black', color='cyan')
plt.title('Distribution of Carbon Dioxide (co2) Values')
plt.xlabel('Carbon Dioxide Values')
plt.ylabel('Frequency')
plt.show()

# Check skewness for the `co2` distribution
if data['co2'].skew() > 0:
    print("The distribution is positively skewed.")
elif data['co2'].skew() < 0:
    print("The distribution is negatively skewed.")
else:
    print("The distribution is symmetric.")




```

```python
# 2. Histogram for `city` - City Fuel Efficiency
plt.figure(figsize=(10, 6))
plt.hist(data['city'], bins=50, edgecolor='black', color='orange')
plt.title('Distribution of City Fuel Efficiency')
plt.xlabel('City Fuel Efficiency Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Histogram for `highway` - Highway Fuel Efficiency
plt.figure(figsize=(10, 6))
plt.hist(data['highway'], bins=50, edgecolor='black', color='green')
plt.title('Distribution of Highway Fuel Efficiency')
plt.xlabel('Highway Fuel Efficiency Values')
plt.ylabel('Frequency')
plt.show()
```

**TO DO:** create a heat map of engine displacement i.e the column "disp"  vs. co2 production.and interpret it (is there correlation?what did you notice?). Engine Displacement is in litres

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/fuel_econ.csv')

# Create a pivot table to structure the data for the heatmap
pivot_table = data.pivot_table(values='co2', index='displ', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='viridis', cbar_kws={'label': 'Average CO2 Production'})
plt.title('Heatmap of Engine Displacement (in Litres) vs. Average CO2 Production')
plt.ylabel('Engine Displacement (L)')
plt.xlabel('CO2 Production')
plt.show()
```
